```python
import queue
class Queue:
    def __init__(self):
        self.queue=queue.Queue(maxsize=4)
    def isempty(self):
        return self.queue.qsize()==0
    def enqueue(self,item):
        try:
            self.queue.put_nowait(item)
            print("Element added")
        except:
            print("Queue size exceeded")
    def getqueue(self):
        if self.isempty():
            print("\n Queue is empty")
            return
        print("Element deleted:",self.queue.get_nowait())
    def size(self):
        return self.queue.qsize()
    def print(self):
         if (self.queue.qsize()==0):
            print("Queue is Empty")
         else:
             print(self.queue.queue)

def main():
    q=Queue()
    while(True):
        print("\n1. Enqueue")
        print("2. Dequeue")
        print("3. Is queue Empty")
        print("4. Size of queue")
        print("5. Print queue")
        print("6. Exit")
        choice=int(input("\nEnter your choice: "))
        if choice==1:
            element=input("Enter the element to add in the queue: ")
            q.enqueue(element)

        elif choice==2:
            q.getqueue()
        elif choice==3:
            print("Is the queue empty: ",q.isempty())
        elif choice==4:
            if q.isempty():
                print("Queue is empty")
            else:
                print("size of the queue:",q.size())
        elif choice==5:
            q.print()
        elif choice==6:
            print("Thank you")
            break
        else:
            print("Invalid input")
```

```python
main()
```
