```python
class Node:
    __slots__ = 'element' , 'prev' , 'next'

    def __init__(self,element,prev,next):
        self.element=element
        self.prev=prev
        self.next=next

class DoublyLinkedList(Node):

    def __init__(self):
        self.header=Node(None,None,None)
        self.trailer=Node(None,None,None)
        self.header.next=self.trailer
        self.trailer.prev=self.header
        self.size=0

    def __len__(self):
        return self.size

    def isempty(self):
        return self.size==0

    def insert_between(self, e, pred,succ):
        new =Node(e,pred,succ)
        pred.next =new
        succ.prev =new
        self.size+=1
        return new

    def delete_node(self,node):
        pred = node.prev
        succ = node.next
        pred.next =succ
        succ.prev =pred
        self.size-=1
        key=node.element
        return key
```

```python
class Deque(DoublyLinkedList):
    def first(self):
        if self.isempty():
            print("Deque is Empty!")
        else:
            front=self.header.next.element
            print("The front element is",front)

    def last(self):
        if self.isempty():
            print("Deque is Empty!")
        else:
            rear=self.trailer.prev.element
            print("The rear element is",rear)

    def insert_first(self,element):
        self.insert_between(element,self.header,self.header.next)

    def insert_last(self,element):
        self.insert_between(element,self.trailer.prev, self.trailer)

    def delete_first(self):
        if self.isempty():
            print("Deque is Empty!")
        else:
            a=self.delete_node(self.header.next)
            print("The deleted element is ",a)

    def delete_last(self):
        if self.isempty():
            print("Deque is Empty!")
        else:
            element=self.delete_node(self.trailer.prev)
            print("The deleted element is ",element)


    def display(self):
            if self.isempty():
                print("The deque is empty")
            else:
                temp =self.header
                while temp:
                    print(temp.element,end=" ")
                    temp =temp.next
```

```python
def main():
    D = Deque()
    while True:
        print("\n\nOperations:\n1.Insert at beginning\n2.Insert at End\n3.Delete at Front\n4.Delete at Rear")
        print("5.Front\n6.Rear\n7.Display\n8.IsEmpty\n9.Size\n10.Exit")
        choice = int(input("\nChoose the operation : "))

        if choice ==1:
            front = input("Enter the element to be added: ")
            D.insert_first(front)
            D.display()

        elif choice == 2:
            item = input("Enter the element to be added: ")
            D.insert_last(item)
            D.display()

        elif choice==3:
            D.delete_first()

        elif choice==4:
            D.delete_last()

        elif choice==5:
            D.first()

        elif choice==6:
            D.last()

        elif choice==7:
            D.display()

        elif choice==8:
            if D.isempty():
                print("The Deque is empty")
            else:
                print("The Deque is not empty")

        elif choice==9:
            if D.isempty():
                print("Size of Deque is 0")
            else:
                print("Size of Deque is: ", len(D))

        elif choice == 10:
            print("END")
            break

        else:
            print("Invalid Choice")
```
