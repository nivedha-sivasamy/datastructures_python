```python
import numpy as np
import sys

class Graph:
  def __init__(self,vertices):
    self._vertices=vertices
    self._adjMat=np.zeros((vertices,vertices))

  def insert_edge(self,u,v,x=1):
    self._adjMat[u][v]=x

  def edges(self):
    for i in range(self._vertices):
      for j in range(self._vertices):
        if self._adjMat[i][j]!=0:
          print(i,'---',self._adjMat[i][j],'--->',j)

  def vertices(self):
    for i in range(self._vertices):
      print(i,end=' ')
    print()

  def display_adjMat(self):
    print(self._adjMat)

  def printsolution(self,dist):
    print("Vertex \t Distance from Source")
    for node in range (self._vertices):
      print(node,"\t",dist[node])

  def minDistance(self,dist,sptSet):
    min=sys.maxsize
    for u in range(self._vertices):
      if dist[u]<min and sptSet[u]==False:
        min=dist[u]
        min_index=u
    return min_index


  def dijkstra(self,src):
    dist=[sys.maxsize]*self._vertices
    dist[src]=0
    sptSet=[False]*self._vertices
    for cout in range(self._vertices):
      x=self.minDistance(dist,sptSet)
      sptSet[x]=True
      for y in range(self._vertices):
        if self._adjMat[x][y]>0 and sptSet[y]==False and dist[y]>dist[x]+self._adjMat[x][y]:
          dist[y]=dist[x]+self._adjMat[x][y]
    self.printsolution(dist)


def main():
  n=int(input("Enter the number of vertices : "))
  g=Graph(n)
  g.vertices()
  while True:
    print("\n************************************\n")
    print("The options are")
    print("1.To Insert Edge\n2.Edges\n3.To Find Shortest Path\n4.Exit")
    choice=int(input("Enter your choice "))
    if choice==1:
      print("To Insert Edge")
      numbr=int(input("Enter the number of edges : "))
      for i in range(numbr):
        v1=int(input("Enter the Vertex : "))
        v2=int(input("Enter the opposite Vertex : "))
        w=int(input("Enter the weight : "))
        g.insert_edge(v1,v2,w)
      g.display_adjMat()

    elif choice==2:
      print("The Edges are : ",g.edges())

    elif choice==3:
      s=int(input("Enter the starting vertex for the shortest path : "))
      g.dijkstra(s)

    else:
      break
```

```python
main()
```
