# Linked List Program

```python

class Node:
    __slots__ = 'data','next'
    def __init__(self,data,next):
        self.data=data
        self.next=next

class Linkedlist:
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0

    def lsize(self):
        return self.size

    def isempty(self):
        return self.size==0

    def insert_at_last(self,element):
        temp=Node(element,None)
        if self.head == None:
            self.head=temp
            self.size+=1
            return
        last=self.head
        while (last.next) :
            last=last.next
        last.next = temp
        self.size+=1

    def display(self):
        temp=self.head
        while temp:
            print(temp.data,end=" ---> ")
            temp=temp.next
        print("NULL")

    def search(self,key):
        temp=self.head
        index=0
        a=[]
        while temp:
            if temp.data.upper()==key.upper():
                a.append(index)
            temp=temp.next
            index+=1
        return a

    def insert_at_position(self,key,position):
        if position == 0:
            self.insert_at_start(key)
        elif position == self.size:
            self.insert_at_last(key)
        elif position<0 or position>self.size:
            print("The position is out of range")
        else:
            new=Node(key,None)
            p=self.head
            index=0
            while index<position-1:
                p=p.next
                index+=1
            new.next=p.next
            p.next=new
            self.size+=1


    def insert_at_start(self,key):
        new=Node(key,None)
        new.next=self.head
        self.head=new
        self.size+=1

    def delete(self,key):
        temp = self.head
        if self.isempty():
            print("List is empty")
        else:
            if (temp is not None): # remove
                if (temp.data == key):
                    self.head = temp.next
                    temp = None
                    self.size-=1
                    return
        while(temp is not None):
            if temp.data == key:
                break
            prev = temp
            temp = temp.next
        if(temp == None):
            print("Element not present in list")
            return
        self.size-=1
        prev.next = temp.next
        temp = None

    def search1(self,key):
        temp=self.head
        index=0
        a=[]
        while temp:
            if temp.data==key:
                a.append(index)
            temp=temp.next
            index+=1
        return a

```

```python
def main():
    l=Linkedlist()
    print("This Program Creates a Linked List Data Structure")
    print("Create a list")
    i=input("Enter the element to be added ")
    l.insert_at_last(i)
    l.display()
    while True:
        print("\nThe operations available are")
        print("1.Insert at end\n2.Search\n3.Display\n4.Insert at position\n5.Insert at beginning\n6.Size\n7.Delete\n8.Is Empty\n9.Exit")
        c=input("Enter Your choice ")

        if c in ("1",'2','3','4','5','6','7','8','9'):
            if c=='1':
                i=input("Enter the element to be added ")
                l.insert_at_last(i)
                l.display()

            if c=='2':
                print("The position is based on index 0")
                if l.isempty():
                    print("List is Empty")
                else:
                    i=input("Enter the element to be searched ")
                    j=l.search(i)
                    if len(j)==0:
                        print("Element is not present in list")
                    else:
                        print("The element",i,"is in position ",j)

            if c=='3':
                if l.isempty():
                    print("List is Empty")
                else:
                    print("The linked list is ")
                    l.display()

            if c=='4':
                print("The position is based on index 0")
                i=input("Enter the element to be added ")
                j=int(input("Enter the position for the element "))
                l.insert_at_position(i,j)
                l.display()

            if c=='5':
                i=input("Enter the element to be added ")
                if l.isempty():
                    l.insert_at_last(i)
                    l.display()
                else:
                    l.insert_at_start(i)
                    l.display()

            if c=='6':
                if l.isempty():
                    print("The size of list is 0")
                else:
                    print("The size of list is",l.lsize())

            if c=='7':
                if l.isempty():
                    print("List is Empty")
                else:
                    key=input("Enter the element to be deleted ")
                    a=len(l.search1(key))
                    if a>1:
                        for i in range(a):
                            l.delete(key)
                        l.display()
                    else:
                        l.delete(key)
                        l.display()

            if c=='8':
                if l.isempty():
                    print("List is empty")
                else:
                    print("List is not empty")


            if c=='9':
                print("End")
                break

        else:
            print("Invalid Choice")
```

```python
main()
```
