```python
def main():
    from collections import deque
    s=deque()
    print("This Program Creates a Stack Data Structure")
    print("The operations available are")
    print("1.Push\n2.Pop\n3.Top\n4.Size\n5.Is Empty\n6.Display\n7.Exit")
    while True:
        c=input("Enter Your choice ")
        if c in ("1",'2','3','4','5','6','7'):
            if c=='1':
                i=input("Enter the element to be added")
                s.append(i)
                print(s)
            if c=='2':
                if len(s)==0:
                    print("Stack is empty")
                else:
                    print("The removed element is ",s.pop())
            if c=='3':
                if len(s)==0:
                    print("Stack is empty")
                else:
                    print("The Top most element is",s[-1])
            if c=='4':
                if len(s)==0:
                    print("Stack is empty")
                else:
                    print("The no of elements is",len(s))
            if c=='5':
                if (len(s)==0):
                    print("True")
                else:
                    print("False")
            if c=='6':
                if len(s)==0:
                    print("Stack is empty")
                else:
                    print(s)
            if c=='7':
                print("End")
                break
        else:
            print("Invalid choice")
```

```python
main()
```
