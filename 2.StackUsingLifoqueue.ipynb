```python
from queue import LifoQueue

class QStack:
    def __init__(self):
        self.values = LifoQueue(maxsize=2)

    def push(self, value):
        if self.values.qsize()>100:
            print("Stack Overflow")
        else:
            return self.values.put(value)

    def pop(self):
        if self.values.empty():
            print("Trying to delete elements from empty stack! Please add elements")
        else:
            return self.values.get()

    def top(self):
       return self.values[-1]

    def empty(self):
        return self.values.empty()

    def size(self):
        return self.values.qsize()


```

```python
def main():
    QS = QStack()
    while True:
        try:
            print("The options are :\n1.Push\n2.Pop\n3.Top\n4.Check Empty Stack\n5.Size of Stack\n6.Exit")
            choice  =  int(input("\nChoose the operation you want to perform: "))
            if choice ==1:
                item = input("\nEnter the elements to be added: ")
                QS.push(item)
            elif choice ==2:
                print("The element popped: ", QS.pop())
            elif choice ==3:
                if QS.empty():
                    print("Empty Stack!!")
                else:
                    print("The first element of Stack: ",QS.top())
            elif choice ==4:
                print("Stack Empty?",  QS.empty())
            elif choice ==5:
                print("The number of elements in stack is:", QS.size())
            elif choice == 6:
              print("Exit")
              break
            else:
                print("Invalid choice")
        except ValueError:
            print("Not a valid choice!")
            break
```

```python
main()
```
