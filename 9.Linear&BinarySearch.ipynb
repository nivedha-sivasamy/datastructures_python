```python
def Binary_Search(Array,key,n):
    L=0
    U=n-1
    mid=0
    while L<=U:
        mid=(L+U)//2
        if Array[mid]==key:
            print("Element found at position",mid+1)
            return
        elif Array[mid]<key:
            L=mid+1
        else:
            Array[mid]>key
            U=mid-1
    print("Element not found")
    return

def Binary_Search_Recursion(Array,n,key,L,U):
    mid=(L+U)//2
    if L>U:
        print("Element not found")
        return
    elif Array[mid]==key:
        print("Element is found at position",mid+1)
        return
    elif Array[mid]<key:
        Binary_Search_Recursion(Array,n,key,mid+1,U)
    else:
        Array[mid]>key
        Binary_Search_Recursion(Array,n,key,L,mid-1)

def Linear_Search(Array,key):
    n=len(Array)
    index=0
    while index<n:
        if Array[index]==key:
            print("The element is at position",index+1)
            return
        index=index+1
    print("Element not found")
    return

def main():
    import numpy as np
    lst=[]
    numbr=int(input("Enter the number of elements you want to enter: "))
    for i in range(numbr):
        element=(input("Enter the elements :"))
        lst.append(element)
    array=np.array(lst)
    n=len(array)
    while True:

        print("\nTypes of search")
        print("\n1.Linear Search\n2.Binary Search\n3.Binary Search with Recursion\n4.Exit")
        choice=input("\nEnter your choice of search :")

        if choice=='1':
            print("Linear Search")
            print(array)
            a=(input("\nEnter the element to be searched :"))
            Linear_Search(array,a)

        elif choice=='2':
            print("Binary Search")
            a1=sorted(lst)
            a2=np.array(a1)
            print("Sorted array",a1)
            a=(input("\nEnter the element to be searched :"))
            Binary_Search(a2,a,n)

        elif choice=='3':
            print("Binary Search with recursion")
            a2=np.array(sorted(lst))
            print("Sorted array",a1)
            a=(input("\nEnter the element to be searched :"))
            L=0
            U=n-1
            Binary_Search_Recursion(a2,n,a,L,U)

        elif choice=='4':
            print("END")
            break

        else:
            print("Invalid Choice")



```
